## Problem 3

#### In this problem, you will type out the input lines starting at "Computations with Categoricals" and end before "Better performance with categoricals" of the textbook.

```python
# Insert commenting and code in this cell.
# Computations with categoricals
# random data
np.random.seed(12345)
draws = np.random.randn(1000)
draws[:5]
```

```python
# quartile
bins = pd.qcut(draws, 4)
bins
```

```python
# quartile names
bins = pd.qcut(draws, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
bins
```

```python
bins.codes[:10]
```

```python
# groupby
bins = pd.Series(bins, name='quartile')
results = (pd.Series(draws)
           .groupby(bins)
           .agg(['count', 'min', 'max'])
           .reset_index())
results
```

```python
results['quartile']
```

```python
# Insert commenting and code in this cell.
# Group transforms an unwrapped groupbys
df = pd.DataFrame({'key': ['a', 'b', 'c'] * 4, 'value': np.arange(12.)})
df
```

```python
# group mean by key
g = df.groupby('key').value
g.mean()
```

```python
# transforms
g.transform(lambda x: x.mean())
```

```python
g.transform('mean')
```

```python
g.transform(lambda x: x * 2)
```

```python
# descending order
g.transform(lambda x: x.rank(ascending=False))
```

```python
# normalize function
def normalize(x):
    return (x - x.mean()) / x.std()
g.transform(normalize)
```

```python
g.apply(normalize)
```

```python
# built in functions
g.transform('mean')
```

```python
normalized = (df['value'] - g.transform('mean')) / g.transform('std')
normalized
```

```python
# Grouped Time Resampling
N = 15
times = pd.date_range('2017-05-20 00:00', freq='1min', periods=N)
df = pd.DataFrame({'time': times, 'value': np.arange(N)})
df
```

```python
# time
df.set_index('time').resample('5min').count()
```

```python
# multiple time series
df2 = pd.DataFrame({'time': times.repeat(3), 'key': np.tile(['a', 'b', 'c'], N), 'value': np.arange(N * 3.)})
df2[:7]
```

```python
# Time grouper no longer in pandas using grouper

#time_key = pd.TimeGrouper('5min')
time_key = pd.Grouper(freq='5min')
resampled = (df2.set_index('time').groupby(['key', time_key]).sum())
resampled
```
