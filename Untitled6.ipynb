```python
# Step 2) 1. Time Series Line Plot
import pandas as pd

from matplotlib import pyplot as plt

# Load as a Series (not a 1-col DataFrame)
series = pd.read_csv('daily-min-temperatures.txt',header=0,index_col=0,
                     parse_dates=True).iloc[:, 0]        # <- make it a Series

series.name = 'temp'

# Group by year-end (YE) instead of deprecated 'A'
groups = series.groupby(pd.Grouper(freq='YE'))

# Build a DataFrame with one column per year
years = pd.DataFrame({name.year: grp.values for name, grp in groups})

# Plot
axarr = years.plot(subplots=True, legend=False, figsize=(8, 10), sharex=False)
plt.tight_layout()
plt.show()
```

```python
# Step 3) 2. Time Series Histogram and Density Plots
# make and show histogram
series.hist()
pyplot.show()
```

```python
# density plot
series.plot(kind='kde')
pyplot.show()
```

```python
# Step 4) 3. Time Series Box and Whisker Plots by Interval
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot

# changed to your code
series = pd.read_csv('daily-min-temperatures.txt',header=0,index_col=0,
                     parse_dates=True).iloc[:, 0]

# chaged to YE since A is depricated
groups = series.groupby(Grouper(freq='YE'))
years = DataFrame()
for name, group in groups:
	years[name.year] = group.values
years.boxplot()
pyplot.show()
```

```python
#distribution of values across months within a year.
from pandas import concat
one_year = series['1990']

# reused seris from previous section
# changed to non deprecated
groups = one_year.groupby(Grouper(freq='ME'))
months = concat([DataFrame(x[1].values) for x in groups], axis=1)
months = DataFrame(months)
months.columns = range(1,13)
months.boxplot()
pyplot.show()
```

```python
# Step 5) 4. Time Series Heat Maps
groups = series.groupby(Grouper(freq='YE'))
years = DataFrame()
for name, group in groups:
	years[name.year] = group.values
years = years.T
pyplot.matshow(years, interpolation=None, aspect='auto')
pyplot.show()
```

```python
# months within a year
one_year = series['1990']
groups = one_year.groupby(Grouper(freq='ME'))
months = concat([DataFrame(x[1].values) for x in groups], axis=1)
months = DataFrame(months)
months.columns = range(1,13)
pyplot.matshow(months, interpolation=None, aspect='auto')
pyplot.show()
```

```python
# Step 6) 5. Time Series Lag Scatter Plots
# create a scatter plot
# lag plot for the Minimum Daily Temperatures dataset.
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import lag_plot
# changed to your code
series = pd.read_csv('daily-min-temperatures.txt',header=0,index_col=0,
                     parse_dates=True).iloc[:, 0]
lag_plot(series)
pyplot.show()
```

```python
# multiple scatter plots
from pandas import concat
values = DataFrame(series.values)
lags = 7
columns = [values]
for i in range(1,(lags + 1)):
	columns.append(values.shift(i))
dataframe = concat(columns, axis=1)
columns = ['t+1']
for i in range(1,(lags + 1)):
	columns.append('t-' + str(i))
dataframe.columns = columns
pyplot.figure(1)
for i in range(1,(lags + 1)):
	ax = pyplot.subplot(240 + i)
	ax.set_title('t+1 vs t-' + str(i))
	pyplot.scatter(x=dataframe['t+1'].values, y=dataframe['t-'+str(i)].values)
pyplot.show()
```

```python
# Step 7) 6. Time Series Autocorrelation Plots
# quantify the strength and type of relationship between observations and their lags.
# create an autocorrelation plot
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(series)
pyplot.show()
```
