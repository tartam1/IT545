```python
#Introduction to statsmodels
# estimating linear models
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
rng = np.random.default_rng(seed=12345)

def dnorm(mean, variance, size=1):
    if isinstance(size, int):
        size = size,
    return mean + np.sqrt(variance) * rng.standard_normal(*size)

N = 100
X = np.c_[dnorm(0, 0.4, size=N),
          dnorm(0, 0.6, size=N),
          dnorm(0, 0.2, size=N)]
eps = dnorm(0, 0.1, size=N)
beta = [0.1, 0.3, 0.5]

y = np.dot(X, beta) + eps
```

```python
X[:5]
```

```python
y[:5]
```

```python
# linear model fitted with an intercept term
X_model = sm.add_constant(X)
X_model[:5]
```

```python
# sm.OLS class
model = sm.OLS(y, X)
results = model.fit()
results.params
```

```python
# summary of resultsw
print(results.summary())
```

```python
# statsmodels has returned results as Series with the DataFrame column names
results.predict(data[:5])
```

```python
# Estimating Time Series Processes
init_x = 4

values = [init_x, init_x]
N = 1000

b0 = 0.8
b1 = -0.4
noise = dnorm(0, 0.1, N)
for i in range(N):
    new_x = values[-1] * b0 + values[-2] * b1 + noise[i]
    values.append(new_x)
```

```python
# lags
from statsmodels.tsa.ar_model import AutoReg
MAXLAGS = 5
model = AutoReg(values, MAXLAGS)
results = model.fit()

#estimated parameters in the results have the intercept first
results.params
```
